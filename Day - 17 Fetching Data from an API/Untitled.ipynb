```python

import time

dfs = []

for i in range(1, 429):
    try:
        response = requests.get(
            "https://api.themoviedb.org/3/movie/top_rated",
            params={
                "api_key": "bf18e20ee7a5ad7f51e4134400bfa654",
                "language": "en-US",
                "page": i
            },
            timeout=10
        )

        if response.status_code != 200:
            print(f"Skipped page {i} | Status {response.status_code}")
            continue

        data = response.json()["results"]

        temp_df = pd.DataFrame(data)[[
            "id",
            "title",
            "overview",
            "release_date",
            "popularity",
            "vote_average",
            "vote_count"
        ]]

        dfs.append(temp_df)

        print(f"Fetched page {i}")

        time.sleep(0.3)  # ðŸ”´ THIS IS CRITICAL

    except Exception as e:
        print(f"Error on page {i}: {e}")
        time.sleep(1)

df = pd.concat(dfs, ignore_index=True)
df
```
