```python
import requests

url = "https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page=1"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.9",
    "Accept-Encoding": "gzip, deflate, br",
    "Connection": "keep-alive",
    "Upgrade-Insecure-Requests": "1",
    "Referer": "https://www.ambitionbox.com/"
}

webpage = requests.get(url, headers=headers)
webpage = webpage.text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
company = soup.find_all('div', class_ = 'companyCardWrapper')
```

```python
len(company)
```

```python
name = []
rating = []

for i in company:
    name.append(i.find('h2').get_text(strip=True))

    reviews = i.find_all('p', class_='rating_text')
    if reviews:
        rating.append(reviews[0].get_text(strip=True))
    else:
        rating.append(None)
```
