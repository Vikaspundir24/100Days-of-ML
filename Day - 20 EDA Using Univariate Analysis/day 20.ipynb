# 1. Categorical Data

## a.CountPlot


```python
sns.countplot(x  = 'Survived', data = df,palette='Set1', hue='Survived')
#df['Survived'].value_counts().plot(kind = 'bar')
```

```python
sns.countplot(x  = 'Sex', data = df, palette='Set1', hue='Sex')
```

```python
sns.countplot(x  = 'Pclass', data = df,palette='Set1', hue='Pclass')
```

```python
sns.countplot(data = df, x = 'Embarked', palette = 'Set1', hue = 'Embarked')
```

```python
df['Survived'].value_counts().plot(kind = 'pie', autopct = '%.2f')
```

```python
df['Sex'].value_counts().plot(kind = 'pie', autopct = '%.2f')
```
